#### creation of deep learnig model
- Step1. Creating a simple Keras Model for IRIS dataset



```python
#import libraries
import numpy as np
import pandas as pd
iris = pd.read_csv("Data/Iris.csv")
iris.head()
```

```python
iris.columns
```

#### Step : 2 definig x and y for training purpose


```python
x =iris.drop('Species',axis=1)
y = iris['Species']
```

```python
x
```

```python
x.shape
```

```python
y
```

#### step 3: Binarizing the labels of species

- encode the target values into numbers
- for doing this we use LabelBinarizer from sklearn.preprocessing library

```python
from sklearn.preprocessing import LabelBinarizer
encoder = LabelBinarizer()
Y = encoder.fit_transform(y)
```

```python
Y
```

#### step 4: Scaling the data
- it is always important to scale our data before feeding the model to ensure 
- we have the values of all columns present in the same range
- scalers present in scikit-learn
- StandardScaler or MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler_x = scaler.fit_transform(x)
```

```python
scaler_x
```

```python
!pip install tensorflow
```

#### step 5. Creating keras model

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units = 4, activation ='relu'))
model.add(Dense(units=32,activation='relu'))
model.add(Dense(units=16,activation='relu'))
# Last layer for multi-class classification of 3 species
model.add(Dense(units=3,activation='softmax'))
```

```python
model.compile(optimizer ='adam',loss="categorical_crossentropy",metrics=['accuracy'])
```

#### Step 6. Training and saving model

```python
#implementing earlystopping
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
model.fit(scaler_x,Y,epochs=200,callbacks=[callback])
```

```python
# saving our model
model.save("final_iris_model.h5")
```
